```python
# Finding the latitude and longitude of San Francisco
address = 'San Francisco, California'
geolocator = Nominatim(user_agent="sfo_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude

# Finding popular places in San Francisco using Foursquare API
url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            latitude, 
            longitude)
results = requests.get(url).json()['response']['groups'][0]['items']
venue_names=[]
venue_categories=[]
for dictionaries in  results:
    venue_names.append(dictionaries['venue']['name'])
    venue_categories.append(dictionaries['venue']['categories'][0]['name']) 
popular_places=pd.DataFrame(list(zip(venue_names,venue_categories)),columns=['Name','Category'])
popular_places
```

```python
# read in climate data
climate=pd.read_csv("Climate.csv",na_values=('T','-'))
# replace missing values with 0
climate=climate.fillna(0)
climate['Day'], climate['Month'] = climate['Day'].str.split('-', 1).str
sfo=climate[climate['City']=='San Francisco']
## December has only 14 days of data
la=climate[climate['City']=='Los Angeles']
## December has only 21 days of data
## November has only 23 days of data
sd=climate[climate['City']=='San Diego']
## December has only 26 days of data
sj=climate[climate['City']=='San Jose']
## December has only 26 days of data
```

```python
# calculate average monthly high, low temperatures and rainfall
climate=climate.iloc[:,[0,2,3,4,7]]
climate = climate.astype({"High(degree F)": float, "Low(degree F)": float,"Precip. Inch":float})
climate=climate.groupby(['City','Month']).mean().reset_index()
month_dict={'Jan':1,'Feb':2,'Mar':3,'Apr':4,'May':5,'Jun':6,'Jul':7,'Aug':8,'Sep':9,'Oct':10,'Nov':11,'Dec':12}
climate['Month_Number']= climate['Month'].map(month_dict)
climate=climate.sort_values(by=['City','Month_Number'])
```

```python
# create plot of variation in high and low temperatures throughout the year
plt.figure(figsize=(16,8))
plt.plot('Month','High(degree F)',data=climate[climate['City']=='San Francisco'],color='blue',label='San Francisco High')
plt.plot('Month','High(degree F)',data=climate[climate['City']=='Los Angeles'],color='red',label='Los Angeles High')
plt.plot('Month','High(degree F)',data=climate[climate['City']=='San Jose'],color='green',label='San Jose High')
plt.plot('Month','High(degree F)',data=climate[climate['City']=='San Diego'],color='cyan',label='San Diego High')
plt.plot('Month','Low(degree F)','--',data=climate[climate['City']=='San Francisco'],color='blue',label='San Francisco Low')
plt.plot('Month','Low(degree F)','--',data=climate[climate['City']=='Los Angeles'],color='red',label='Los Angeles Low')
plt.plot('Month','Low(degree F)','--',data=climate[climate['City']=='San Jose'],color='green',label='San Jose Low')
plt.plot('Month','Low(degree F)','--',data=climate[climate['City']=='San Diego'],color='cyan',label='San Diego Low')
plt.title('Average monthly low and high temperatures',fontsize='x-large')
plt.xlabel("Months",fontsize='x-large')
plt.ylabel("Temperature (degree F)",fontsize='x-large')
plt.xticks(fontsize='large')
plt.legend()
```

```python
# create plot of variation in rainfall throughout the year
plt.figure(figsize=(12,6))
plt.plot('Month','Precip. Inch',data=climate[climate['City']=='San Francisco'],color='blue',label='San Francisco')
plt.plot('Month','Precip. Inch',data=climate[climate['City']=='Los Angeles'],color='red',label='Los Angeles')
plt.plot('Month','Precip. Inch',data=climate[climate['City']=='San Jose'],color='green',label='San Jose')
plt.plot('Month','Precip. Inch',data=climate[climate['City']=='San Diego'],color='cyan',label='San Diego')
plt.title('Average monthly rainfall (inches)',fontsize='x-large')
plt.xlabel("Months",fontsize='x-large')
plt.ylabel("Rainfall (inches)",fontsize='x-large')
plt.xticks(fontsize='large')
plt.legend()
```

```python
# Read in crime data
crime=pd.read_excel('Crime.xlsx')

# Filter crime data for Los Angeles, San Francisco, San Jose and San Diego
crime=crime[crime['geoname'].str.contains('Los Angeles|San Jose|San Francisco|San Diego', case=False,na=False)]
city_dict={'Los Angeles':'Los Angeles','San Diego':'San Diego','San Francisco':'San Francisco','East Los Angeles CDP':'Los Angeles','Lake Los Angeles CDP':'Los Angeles','Los Angeles city':'Los Angeles','Rancho San Diego CDP':'San Diego','San Diego city':'San Diego','San Diego Country Estates':'San Diego',
           'San Francisco city':'San Francisco','San Jose city':'San Jose','South San Francisco city':'San Francisco','South San Jose Hills CDP':'Los Angeles'}
crime['City']= crime['geoname'].map(city_dict)
crime=crime[crime['strata_level_name_code']==5]
crime=crime[['reportyear','City','numerator','denominator']]
crime=crime.groupby(['reportyear','City']).sum().reset_index()

# Calculate crime rate for each city
crime['rate']=crime['numerator']/crime['denominator']*1000
```

```python
# create plot of change in crime rate over the years from 2000 to 2013
plt.figure(figsize=(12,6))
plt.plot('reportyear','rate',data=crime[crime['City']=='San Francisco'],color='blue',label='San Francisco')
plt.plot('reportyear','rate',data=crime[crime['City']=='Los Angeles'],color='red',label='Los Angeles')
plt.plot('reportyear','rate',data=crime[crime['City']=='San Jose'],color='green',label='San Jose')
plt.plot('reportyear','rate',data=crime[crime['City']=='San Diego'],color='cyan',label='San Diego')
plt.title('Violent crimes per 1000 population from 2000 to 2013',fontsize='x-large')
plt.xlabel("Year",fontsize='x-large')
plt.ylabel("Crime rate per 1000 population",fontsize='x-large')
plt.xticks(fontsize='large')
plt.legend()
```
